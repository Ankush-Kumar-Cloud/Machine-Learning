```python
titanic = pd.read_csv('train.csv')
titanic.sample(5)
```

### Barplot (Numerical-Categorical)

```python
sns.barplot(x='Pclass' , y='Age' , data = titanic)
```

```python
sns.barplot(x='Pclass' , y='Fare' , data = titanic , hue = 'Sex')
```

### Boxplot (Numerical-Categorical)

```python
sns.boxplot(x='Sex',y='Age' , data = titanic , hue = 'Survived' , saturation=0.75)
```

### Distplot(Numerical-Catigorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

### Heatmap (Numerical-Categorical)

```python
titanic.head(5)
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
plt.show()
```

```python
titanic.groupby('Sex')['Survived'].sum().plot(kind='pie',autopct="%0.1f%%")
```

```python
sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived']))
plt.show()
```

### lineplot (Numerical-Numerical)

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
new = flights.groupby('year')['passengers'].sum().reset_index()
```

```python
sns.lineplot(x='year', y = 'passengers' , data = new)
```

```python
flights.pivot_table(values='passengers',index = 'month' , columns = 'year')
```

```python
sns.heatmap(flights.pivot_table(values='passengers',index = 'month' , columns = 'year'))
```

```python
sns.clustermap(flights.pivot_table(values='passengers',index = 'month' , columns = 'year'))
```
